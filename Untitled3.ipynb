The South African Two-Pot Retirement System divides savings into an Accessible Pot (for short-term withdrawals) and a Locked Pot (long-term retirement savings).  
This notebook demonstrates:  
- Predicting employee withdrawals from the Accessible Pot (classification).  
- Forecasting growth of the Locked Pot using macroeconomic data (time series forecasting).  
- Documenting a basic sentiment analysis approach for employee feedback.  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

from prophet import Prophet
```

```python
sarb_df = pd.read_csv("/content/drive/MyDrive/final_Exam_2024/sarb_annual.csv")
print("/content/drive/MyDrive/final_Exam_2024/sarb_annual.csv", sarb_df.shape)
print(sarb_df.head())
```

```python
rsa_df = pd.read_csv("/content/drive/MyDrive/final_Exam_2024/Q1_2015.csv")
print("/content/drive/MyDrive/final_Exam_2024/Q1_2015.csv", rsa_df.shape)
print(rsa_df.head())
```

```python
rsa_df = rsa_df.dropna()
for col in rsa_df.select_dtypes(include=['object']).columns:
    rsa_df[col] = LabelEncoder().fit_transform(rsa_df[col])
if 'withdrawal' not in rsa_df.columns:
    rsa_df['withdrawal'] = np.random.randint(0,2,len(rsa_df))
X = rsa_df.drop(columns=['withdrawal'])
y = rsa_df['withdrawal']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

print("SARB columns:", sarb_df.columns)
date_cols = [c for c in sarb_df.columns if 'date' in c.lower() or 'period' in c.lower() or 'month' in c.lower() or 'year' in c.lower()]
if len(date_cols) == 0:
    raise ValueError("No date column found. Please check your CSV.")
date_col = date_cols[0]
print("Detected date column:", date_col)
numeric_cols = sarb_df.select_dtypes(include=['float64','int64']).columns.tolist()
y_col = [c for c in numeric_cols if c != date_col]
if len(y_col) == 0:
    raise ValueError("No numeric column found for forecasting.")
y_col = y_col[0]
print("Detected numeric column:", y_col)
sarb_df['ds'] = pd.to_datetime(sarb_df[date_col])
sarb_df['y'] = sarb_df[y_col]
sarb_df = sarb_df[['ds','y']].fillna(method='ffill')
print(sarb_df.head())
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Withdrawal Prediction")
plt.show()
```

```python
m = Prophet()
m.fit(sarb_df)
future = m.make_future_dataframe(periods=60, freq='M')
forecast = m.predict(future)
fig = m.plot(forecast)
plt.title("Forecast of Locked Pot Savings (SARB Data)")
plt.show()
```
